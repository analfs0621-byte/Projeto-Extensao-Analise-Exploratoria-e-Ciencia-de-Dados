###📁Tipos de Variáveis:

- **Variável qualitativa nominal:** seus valores possíveis são diferentes categorias não-ordenadas, em que cada observação pode ser classificada.
Exemplos: raça, nacionalidade, área de atividade.

- **Variável qualitativa ordinal:** seus valores possíveis são diferentes categorias ordenadas, em que cada observação pode ser classificada.
Exemplos: classe social, nível de instrução.

- **Variável quantitativa discreta:** seus valores possíveis são em geral resultados de um processo de contagem.
Exemplos: número de filhos, número de séries escolares cursadas com aprovação.

- **Variável quantitativa contínua:** seus valores possíveis podem ser expressos através de números reais e varrem uma escala contínua de medição.
Exemplos: renda mensal, peso, altura.

# 5 📈Estatísticas descritivas

O método `describe()` é utilizado no pandas para gerar um resumo estatístico descritivo das colunas de um DataFrame ou Series.
Ele fornece informações importantes para entender a distribuição dos dados.

Esse método se aplica para variáveis numéricas, o describe() normalmente calcula:

count → quantidade de valores não nulos

mean → média

std → desvio padrão

min → valor mínimo

25% → primeiro quartil

50% → mediana

75% → terceiro quartil

max → valor máximo

Essas medidas ajudam a entender a distribuição dos dados.

```python
df_novo.describe()
```

🔹 Variáveis Numéricas

Ele calcula estatísticas matemáticas completas.

Exemplo: Idade, Salário e Altura



🔹 Variáveis Categóricas (texto)

Se usar: `df.describe(include='object')`

Ele mostra:

count → quantidade de valores

unique → quantidade de categorias únicas

top → categoria mais frequente

freq → frequência da categoria mais comum


### 📐Significado das principais medidas

📌 Média (Mean)

É o valor médio dos dados.

Fórmula:
Soma de todos os valores ÷ quantidade de valores.

Representa o ponto central dos dados.


📌 Mediana (50%)

É o valor que fica no meio quando os dados estão ordenados.

Metade dos valores fica acima e metade abaixo.

É menos afetada por valores extremos (outliers) do que a média.


📌 Desvio Padrão (Std)

Mede o quanto os dados variam em relação à média.

Baixo → dados próximos da média

Alto → dados muito espalhados

Indica a dispersão dos dados.


📌 Mínimo (Min)

Menor valor encontrado no conjunto de dados.


📌 Máximo (Max)

Maior valor encontrado no conjunto de dados.


📌 Quartis

Os quartis dividem os dados em 4 partes iguais:

Q1 (25%) → 25% dos dados estão abaixo desse valor

Q2 (50%) → Mediana

Q3 (75%) → 75% dos dados estão abaixo desse valor

No describe() aparecem como:

25%   → Q1
50%   → Mediana
75%   → Q3


📌 Percentis

Percentis dividem os dados em 100 partes.

Exemplo:

Percentil 10 → 10% dos valores são menores que ele

Percentil 90 → 90% dos valores são menores que ele

Quartis são apenas percentis específicos:

Q1 = Percentil 25

Q2 = Percentil 50

Q3 = Percentil 75


📌 Outliers (Valores Atípicos)

Outliers são valores muito diferentes do padrão dos dados.

Exemplo:

Salários: 2000, 2500, 2700, 3000, 50000

O 50000 é um outlier.

Eles podem ocorrer por: Erros de coleta, Casos raros reais e Problemas no sistema.

**Calculando Outliers**

```python
Q1 = df['salary_in_usd'].quantile(0.25)
Q3 = df['salary_in_usd'].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df['salary_in_usd'] < limite_inferior) |
              (df['salary_in_usd'] > limite_superior)]
```

```python
# Quantos registros existem no DataFrame de outliers? / shape retorna linhas e colunas, aqui estou pegando apenas o número de linhas
outliers.shape[0]
```

-158 registros foram considerados outliers.

-158 pessoas no dataset possuem salários acima do limite estatístico.



```python
Total = 9355
Outliers = 158

resultado = Outliers / Total
round(resultado, 2)
```

```python
# Arredondar em porcentagem

porcentagem = (Outliers / Total) * 100
print(f"{porcentagem:.2f}%")
```

Cerca de 1.69% dos dados são valores extremos

```python
# Calculando os limites pelo método do IQR:

Q1 = df["salary_in_usd"].quantile(0.25)
Q3 = df["salary_in_usd"].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df["salary_in_usd"] < limite_inferior) |
              (df["salary_in_usd"] > limite_superior)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de histograma

plt.figure(figsize=(10,5))

sns.histplot(df["salary_in_usd"], bins=50, kde=True)

plt.axvline(limite_inferior, color="red", linestyle="--", label="Limite Inferior")
plt.axvline(limite_superior, color="red", linestyle="--", label="Limite Superior")

plt.legend()
plt.title("Histograma com Limites de Outliers")

plt.show()
```

```python
total = df.shape[0]
qtd_outliers = outliers.shape[0]

porcentagem = (qtd_outliers / total) * 100

plt.figure(figsize=(8,4))
sns.boxplot(x=df["salary_in_usd"])

plt.title(f"Outliers: {qtd_outliers} ({porcentagem:.2f}%)")

plt.show()

```

Foram identificados 158 registros classificados como outliers, representando aproximadamente 1,69% da base de dados. Esses valores podem indicar salários excepcionalmente altos ou baixos em comparação com a distribuição geral.
